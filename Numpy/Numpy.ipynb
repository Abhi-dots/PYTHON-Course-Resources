```python
# How to create 1D array?
```

```python
arr = np.array([10, 20, 30, 40, 78, 22])

print(arr)
```

```python
# How to create 2D array?
```

```python
students = np.array([[85, 90], [79, 92], [89, 99]])

print(students)
```

```python
# How to create 3D array?
```

```python
data = np.array([[[85, 90], [78, 92]], [[79, 99], [73, 91]]])

print(data)
```
