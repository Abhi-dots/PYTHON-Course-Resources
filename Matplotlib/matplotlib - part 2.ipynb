```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
sales = pd.DataFrame({
    "Month" : ["Jan", "Feb", "Mar", "Apr", "May", "June"],
    "Sales" : [120, 150, 180, 170, 210, 250],
    "Profit" : [20,30,35,40,50,65],
    "Orders" : [100, 130, 150, 145, 180, 220]
})
```

```python
print(sales)
```

```python
# LINE CHART
```

```python
# A line chart is used to show trends or changes in data over time.
```

```python
plt.figure(figsize = (5,3))
plt.plot(sales["Month"], sales["Sales"], marker="o", color = "red", linestyle ="--")
plt.title("Monthly sales report")
plt.xlabel("Month")
plt.ylabel("Sales (in Lakhs)")
```

```python
# BAR CHART
```

```python
# A bar chart is used to compare values across different  categories.
```

```python
plt.figure(figsize = (5,3))
plt.bar(sales["Month"], sales["Sales"], color = "skyblue", width = 0.6)
```

```python
plt.figure(figsize = (5,3))
plt.barh(sales["Month"], sales["Sales"], color = "skyblue")
```

```python
# HISTOGRAM CHART
```

```python
# Histogram is used to show the distribution of numerical data by grouping values into intervals.
```

```python
import numpy as np
salary = np.array([25000,28000,30000,32000,35000,40000,45000,50000,60000,65000,70000])
```

```python
plt.figure(figsize = (5,3))
plt.hist(salary, bins=5)
plt.show()
```

```python
s = pd.DataFrame({
    "Month":["Jan","Feb","Mar","Apr","May","Jun"],
    "Sales":[120,150,180,170,210,250],
    "Profit":[20,30,35,40,50,65],
    "Orders":[100,130,150,145,180,220],
    "Marketing":[15,18,22,20,25,30]
})

print(s)
```

```python
# scatter plot
```

```python
plt.figure(figsize = (5,3))
plt.scatter(s["Marketing"], s["Sales"])
```

```python
# pie chart
```

```python
plt.pie(s["Sales"], labels = s["Month"], autopct = "%1.1f%%")
```

```python
# multiple line charts
```

```python
plt.figure(figsize = (5, 3))

plt.plot(s["Month"], s["Sales"], marker = "o", label="Sales")

for i , value in enumerate(s["Sales"]):
    plt.text(i, value , str(value), ha = "center", va = "bottom")

plt.plot(s["Month"], s["Profit"], marker = "*", label="Profit")

plt.legend(loc="best")
```
